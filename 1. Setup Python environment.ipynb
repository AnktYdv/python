# What is Python?
Python is an interpreted high level programming language. It is often subclassified as a "scripting language" and is considered similar to languages such as Perl, Tcl, or Ruby. The syntax of Python is loosely inspired by elements of C programming.

The primary difference between a scripting language and a programming language is in their execution â€“ programming languages use a compiler to convert the complete high-level programming language program into machine language before executing, on the other hand, scripting languages use an interpreter to execute a scripting language program by converting one line at a time. Scripting language are often called to execute in REPL environment - *Read-Evaluate-Print Loop*.

Python was created by Guido van Rossum around 1990 who named it in honor of Monty Python.

## Running Python
Python programs are executed by an interpreter. Usually, the interpreter is started by simply typing python into a command shell. You may like to start python interpreter by 
* starting command shell aka terminal.
* cd to your training folder (e.g. "Python training"). This is for this tutorial only and is not essential at all. You can start Python shell from any folder.
* Type "python" and press <Enter>

````python
print("Hello World!")
````

````python
"print" is the most commonly used Python function. Note that we use "print", all in small caps, followed by "()" and within 
"()", we specify we mention the arguments to be printed. Here "Hello World" is a string or alphanumeric arguments, hence 
mentioned within double quotes, although we can use single quotes also. 

For numeric arguments, we do not use quotes.
````
